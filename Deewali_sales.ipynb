```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Diwali Sales Data.csv", encoding="unicode_escape")
```

```python
df.shape
# telling about how many rows and columns are in data
```

```python
df.info()
```

```python
# to drop unrelated/null columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
# checking for null values in data
pd.isnull(df)
```

```python
# checking for total null  values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# lets change amount data type to amount
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
# chnging column name
df.rename(columns = {'Marital_Status':'Shadi'})
```

```python
df.describe()
```

```python
# targeting the particular columns
df[['Age','Amount']].describe()
```

```python
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# checking amount of gender mean how much males and females are there
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum()
print(sales_gen)
```

```python
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

From the above graph we can observe that women have high purchasing power then men

```python
df.columns
```

```python
ax=sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_record=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_record)
```

Women between age 26-35 buy more

```python
# on the basis of state
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders', data=sales_state)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount', data=sales_state)
```

Product Category
